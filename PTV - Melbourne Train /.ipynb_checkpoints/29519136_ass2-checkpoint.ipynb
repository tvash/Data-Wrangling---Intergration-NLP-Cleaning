### - Date - correction

```python
def correct_datetime(date):
    r = '(04|06|09|11)'
    date_split = date.split('-')
    if date_split[1] == '02':
        return date_split[0]+'-'+date_split[1]+'-'+'28'
    if re.match(r,date_split[1]):
        return date_split[0]+'-'+date_split[1]+'-'+'30'
    else:
        return date_split[0]+'-'+date_split[2]+'-'+date_split[1]
        
    
```

```python
x = list(df['Departure Date'])
final_list = []
check_list = [] #Dummy list to check weather it is correct or not


for num,value_ in enumerate(x):
    try:
        datetime.datetime.strptime(value_, '%Y-%m-%d')
        final_list.append(value_)
    except ValueError:
        correct_format = correct_datetime(value_)
        final_list.append(correct_format)
        check_list.append(correct_format)
        
#Adding the correct-time back        
df['Departure Date'] = final_list
```

```python
#CHECKING THE TIME DIFFERENCE AND ALSO THE TIME TRAVEL
def time_category(s1,s2,tdel):
    FMT = '%H:%M:%S'
    tdelta = datetime.datetime.strptime(s1, FMT) - datetime.datetime.strptime(s2, FMT)
    str_input_time = math.floor(tdel)
    return tdelta.seconds == str_input_time
```

```python
def time_category1(s1,s2):
    pt =datetime.datetime.strptime(s1,'%H:%M:%S')
    pt2 =datetime.datetime.strptime(s2,'%H:%M:%S')
    diff = pt - pt2
    return diff.seconds

def cond(s1,s2):
    pt =datetime.datetime.strptime(s2,'%H:%M:%S')
    pt2 =datetime.datetime.strptime(s1,'%H:%M:%S')
    diff = pt - pt2
    bool_v = (diff.seconds/(60*60)) >12
    return bool_v
```

```python
#HOT ENCODER
def date_hot_encoder(date_time):
    date_obj = datetime.datetime.strptime(date_time,'%Y-%m-%d')
    if date_obj.strftime('%A') in ['Sunday','Saturday']:
        return 1
    else:
        return 0
    
def tod_hot_encoder(time_o):
        time_obj = datetime.datetime.strptime(time_o,'%H:%M:%S')
        if datetime.datetime.strptime('11:59:59','%H:%M:%S') >= time_obj >= datetime.datetime.strptime('6:0:0','%H:%M:%S'):
            return 0
        elif datetime.datetime.strptime('20:59:59','%H:%M:%S') >= time_obj >= datetime.datetime.strptime('12:00:00','%H:%M:%S'):
            return 1
        else:
            return 2    
```

```python
def IQR(another):
    q_25,q_75 = np.percentile(another['Residuals'],[25,75])
    interquartile = q_75 - q_25
    first = q_75 + (1.5*interquartile)
    last =  q_25 - (1.5*interquartile)
    return [first,last]
```

```python
#Finding the outliers
sub_another = another[another['Uber Type']==0]
Uber_type0_outleirs=sub_another[(sub_another['Residuals']>IQR(sub_another)[0]) | (another['Residuals']<IQR(sub_another)[1])]

sub_another = another[another['Uber Type']==1]
Uber_type1_outleirs=sub_another[(sub_another['Residuals']>IQR(sub_another)[0]) | (another['Residuals']<IQR(sub_another)[1])]

sub_another = another[another['Uber Type']==2]
Uber_type2_outleirs=sub_another[(sub_another['Residuals']>IQR(sub_another)[0]) | (another['Residuals']<IQR(sub_another)[1])]
         
print(f'THE NUMBER OF OUTLIERS: -')
print('')
print(f'Uber 0 has {len(Uber_type0_outleirs)} outliers')
print(f'Uber 1 has {len(Uber_type1_outleirs)} outliers')
print(f'Uber 2 has {len(Uber_type2_outleirs)} outliers')
```
